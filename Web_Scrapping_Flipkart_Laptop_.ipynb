```python
from bs4 import BeautifulSoup ## The library which is used for web scrapping
import pandas as pd
import requests ## Library for reading the url
```

```python
## url of flipkart laptop page
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=7f6a4f8c-8228-4896-bf0d-c2adacf7cb51'
```

```python
page=requests.get(url) ## reading the url
soup=BeautifulSoup(page.content, 'lxml') ## reading the html code of the page
elements=soup.find_all('a', class_="_1fQZEK") ## finding class on which we will work
```

```python
## finding the names of laptops in a page 
for a in elements:
    name=a.find('div',class_="_4rR01T")
    print(name)
```

```python
## finding diiferent things using different classes and printing it in text format
for a in elements:
    name=a.find('div', class_="_4rR01T")
    features=a.find('ul',class_="_1xgFaf")
    ratings=a.find('div',class_="_3LWZlK")
    price=a.find('div',class_="_30jeq3 _1_WHN1")
    print(name.text.strip())
    print(features.text.strip())
    print(ratings.text.strip())
    print(price.text.strip())
    print()
```

```python
str_name=str(name) ## coverting the name into string for cleaning
cleanname=BeautifulSoup(str_name, 'lxml').get_text()

print(cleanname)
```

```python
import re
```

```python
Laptop_name=[] ## Creating a list 
for a in elements:
    name=a.find('div', class_="_4rR01T")
    str_name=str(name)
    clean = re.compile('<.*?>') ## Compiling the names
    clean2 = (re.sub(clean, '',str_name))
    Laptop_name.append(clean2) ## apeending it into the list created above
```

```python
print(Laptop_name)
```

```python
df_1=pd.DataFrame(Laptop_name) ## creating the Laptop_name list into data frame
```

```python
df_1.head()
```

```python
df_1 = df_1[0].str.split('-', expand=True) ## spliting the names to find different features
df_1.head()
```

```python
df_1['Laptop Name'] = df_1[[0]]
```

```python
ram=df_1[1].str.split(')', expand=True) ## preprocessing for finding the ram
```

```python
ram_n=ram[0].str.split('/', expand=True)
```

```python
ram_n
```

```python
ram=ram_n[[0]]
```

```python
ram=ram[0].str.replace('(','') ## removing the '(' before the ram. This is part of cleaning
```

```python
ram=pd.DataFrame(ram) ## creating ram into data frame
```

```python
ram=ram[0].str.split('GB',expand=True)
```

```python
ram['RAM']=ram[0]
ram=ram.drop([0,1],axis=1)
```

```python
df_1=df_1.drop([0,1,2], axis=1)
```

```python
df_1=df_1['Laptop Name'].str.split(' ', expand=True) 
```

```python
df_1['Brand']=df_1[[0]] ## Finding laptop brands 
```

```python
df_1['Series'] = df_1[[1,2]].agg(' '.join, axis=1)
```

```python
x=[i for i in range(0,9,1)]
df_1=df_1.drop(x,axis=1)
```

```python
df_1.head()
```

```python
Features=[]
for a in elements:
    features=a.find('ul',class_="_1xgFaf")
    str_features=str(features)
    clean_features = re.compile('<.*?>')
    clean2_features = (re.sub(clean_features, '',str_features))
    Features.append(clean2_features)
```

```python
df_2=pd.DataFrame(Features)
```

```python
df_2.head()
```

```python
df_2 = df_2[0].str.split('Processor', expand=True)
df_2.head()
```

```python
df_2['Processor'] = df_2[[0]]
```

```python
Processor=df_2.drop([0,1], axis=1)
```

```python
Processor
```

```python
Price=[]
for a in elements:
    price=a.find('div',class_="_30jeq3 _1_WHN1")
    str_price=str(price)
    Price.append(price.text)
```

```python
df_3=pd.DataFrame(Price)
```

```python
df_3.head()
```

```python
df_3['Price']=df_3[0]
```

```python
df_3=df_3.drop([0], axis=1)
```

```python
df_3=df_3['Price'].str.replace('₹','') ## Removing the ₹ sign before price
```

```python
df_3=pd.DataFrame(df_3)
```

```python
df_3=df_3['Price'].str.replace(',','') ## Removing commas from price
```

```python
df_3=df_3.astype(int) ## converting price into integer type
```

```python
Ratings=[]
for a in elements:
    ratings=a.find('div',class_="_3LWZlK")
    Ratings.append(ratings.text)
```

```python
Ratings=pd.DataFrame(Ratings)
```
